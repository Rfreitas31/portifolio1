## Principais Empresas na America Latina:

```python
america_do_sul = ['Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia', 'Ecuador', 
                  'French Guiana', 'Paraguay', 'Peru', 'Suriname', 'Uruguay', 'Venezuela']
empresas_sulamericanas = dados.query('country in @america_do_sul')
```

```python
empresas_sulamericanas.index = range(empresas_sulamericanas.shape[0])
empresas_sulamericanas
    
```

#### Na América do Sul, o Brasil possui a maior quantidade das maiores empresas do mundo.

### Empresas do Brasil:

```python
brasil = ['Brazil']
empresas_brasil = empresas_sulamericanas.query('country in @brasil')
empresas_brasil.index = range(empresas_brasil.shape[0])
empresas_brasil
```
