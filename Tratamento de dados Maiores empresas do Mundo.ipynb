# As maiores empresas do mundo e seus países de origem:

## Dados Básicos:

```python
import pandas as pd
import numpy as np
dados = pd.read_csv('arquivos/maiores_empresas_forbes.csv')
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (15,8))
```

```python
dados.head(50)
```

```python
dados.info()
```

```python
dados.columns
```

```python
dados.columns = dados.columns.str.strip()
```

```python
list(dados['global company'].drop_duplicates())
```

```python
# Quantidade de dados
dados.shape
```

```python
# Países com mais empresas:
paises = dados.country.value_counts()
visual_mundo = paises.to_frame(name= 'Quantidade_geral')
visual_mundo
```

```python
# Selecione qualquer pais:
paises.Brazil
```

## Principais Empresas na America Latina:

```python
america_do_sul = ['Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia', 'Ecuador', 
                  'French Guiana', 'Paraguay', 'Peru', 'Suriname', 'Uruguay', 'Venezuela']
empresas_sulamericanas = dados.query('country in @america_do_sul')
```

```python
empresas_sulamericanas.index = range(empresas_sulamericanas.shape[0])
visual_empresas_sulamericanas = empresas_sulamericanas.country.value_counts()
visual_empresas_sulamericanas
```

```python
area_g2 = plt.figure()
g2 = area_g2.add_subplot(1, 1, 1)
label = visual_empresas_sulamericanas.index #x
valores = visual_empresas_sulamericanas.values #y
g2.bar(label, valores)
g2.set_title('Rank América do Sul')
```

#### Na América do Sul, o Brasil possui a maior quantidade das maiores empresas do mundo.

### Empresas do Brasil:

```python
brasil = ['Brazil']
empresas_brasil = empresas_sulamericanas.query('country in @brasil')
empresas_brasil.index = range(empresas_brasil.shape[0])
empresas_brasil
```
